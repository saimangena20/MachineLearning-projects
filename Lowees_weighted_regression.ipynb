```python
import numpy as np
import statsmodels.api as sm
```

```python
np.random.seed(0)
X = np.random.rand(100, 1) * 2 - 1
y = np.sin(X) + np.random.normal(0, 0.1, size=(100, 1))
# Fit LOWESS model
lowess = sm.nonparametric.lowess
smoothed=lowess (y.flatten(), X.flatten())
# Plot the results
import matplotlib.pyplot as plt
plt.scatter (X, y, label='Original data')
plt.plot(smoothed[:, 0], smoothed[:, 1], color='red', label='LOWESS')
plt.legend()
plt.show()
```
