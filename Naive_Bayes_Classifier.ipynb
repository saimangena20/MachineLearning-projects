```python
import pandas as pd
```

```python
data=pd.read_csv('/content/titanic.csv')
data.head()
```

```python
data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
data.head()
```

```python
target=data.Survived
inputs=data.drop('Survived',axis='columns')
dummies=pd.get_dummies(inputs.Sex)
dummies.head(5)
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head()
```

```python
inputs.columns[inputs.isna().any()]
pd.Index([], dtype='object')
```

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

# **Splitting the dataset into the Training dataset and Test set**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(inputs)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
