```python
from scipy import special
a=special.exp1(3)
print(a)
```

```python
b=special.exp2(3)
print(b)

```

```python
## trignometric function
c=special.sindg(0)
print(c)
```

```python
d=special.cosdg(90)
print(d)
```

```python
## integration function
from scipy import integrate
help(integrate)
```

```python
i=scipy.integrate.quad(lambda x:special.exp1(x),0,1)
print(i)
```

```python
e=lambda x,y:8*x+6*y
f=lambda x:0
g=lambda x:1
integrate.dblquad(e,0,2,f,g)
```

```python
##linear algebra
from scipy import linalg
a=np.array([[1,2],[3,4]])
b=linalg.inv(a)
print(b)
```

```python
import matplotlib.pyplot as plt
from scipy import interpolate
x=np.arange(5,20)
y=np.exp(x/3.0)
f=interpolate.interp1d(x,y)
x1=np.arange(6,12)
y1=f(x1)
plt.plot(x,y,'o',x1,y1,'--')
plt.show()
```
